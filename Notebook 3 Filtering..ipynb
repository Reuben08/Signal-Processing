```python
%matplotlib inline
from __future__ import division
from scipy import signal 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec

gs = gridspec.GridSpec(2,2)
gs.update( wspace=0.5, hspace=0.5)


fig=plt.figure()
fig.set_size_inches((10,5))

ax = fig.add_subplot(plt.subplot(gs[1,:]))
Ns=25
n= np.arange(Ns)
x = np.cos(np.arange(Ns)*np.pi/3)  #building the signal with a frequency of pi/3
ma_length=3
y= signal.lfilter(np.ones(ma_length)/ma_length,1,x) 
ax.stem(n,x,label='input',basefmt='b-')
ax.plot(n,x,':')
ax.stem(n[ma_length-1:],y[:-ma_length+1],markerfmt='ro',linefmt='r-',label='output')
ax.plot(n[ma_length-1:],y[:-ma_length+1],'r:')
ax.set_xlim(xmin=-1.1)
ax.set_ylim(ymin=-1.1,ymax=1.1)
ax.set_xlabel("n",fontsize=18)
ax.set_xticks(n)
ax.legend(loc=0)
ax.set_ylabel("amplitude",fontsize=18);


```

```python
# A 9 Average filter

fig=plt.figure()
fig.set_size_inches((10,5))

ax = fig.add_subplot(plt.subplot(gs[1,:]))
Ns=25
n= np.arange(Ns)
x = np.cos(np.arange(Ns)*np.pi/3)  #building the signal with a frequency of 2*pi
ma_length=9
y= signal.lfilter(np.ones(ma_length)/ma_length,1,x)  

ax.stem(n,x,label='input',basefmt='b-')
ax.plot(n,x,':')
ax.stem(n[ma_length-1:],y[:-ma_length+1],markerfmt='ro',linefmt='r-',label='output')
ax.plot(n[ma_length-1:],y[:-ma_length+1],'r:')
ax.set_xlim(xmin=-1.1)
ax.set_ylim(ymin=-1.1,ymax=1.1)
ax.set_xlabel("n",fontsize=18)
ax.set_xticks(n)
ax.legend(loc=0)
ax.set_ylabel("amplitude",fontsize=18);
```

Thus we have modified the code and the response for both, a moving averager filter of order 3 and 9 is plotted. Here we can see that the response of the low pass filter is coming out to be as close to the estimated output. The frequency used is pi/3.
