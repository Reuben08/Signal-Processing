```python
# the code has been excecuted by using a different clip limit and the observation is printed.


%matplotlib inline
from skimage import novice, data, io, exposure
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


brainpic = novice.open('braincomplete.png')

brainpicarray = mpimg.imread('braincomplete.png','True')
img_eq = exposure.equalize_hist(brainpicarray)
p2, p98 = np.percentile(brainpicarray, (5, 70))

img_eq2 = exposure.rescale_intensity(brainpicarray, in_range=(p2, p98))

fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
img_adapteq = exposure.equalize_adapthist(brainpicarray, clip_limit=0.09)

print(' As you can see the clip limit has been changed and its current value is 1. By increasing the clip limit, we are actu-ally increasing the contrast of the image ,especially when the usable data of the image is represented by close contrast values. Through this adjustment, the intensities can be better distributed on the histogram. This allows    the areas of lower local contrast to gain a higher contrast')

plt.subplot(221)
plt.imshow(img_adapteq)

plt.subplot(222)
plt.imshow(brainpicarray)
plt.subplot(223)
ht=plt.hist(img_eq2.ravel(), bins=256, histtype='step', color='black')
plt.subplot(224)
ht=plt.hist(brainpicarray.ravel(), bins=256, histtype='step', color='black')
```

### The syntax used to implement the function equalize_hist is :           img_eq = exposure.equalize_hist(brainpicarray) 

###  The parameter use in this function is ' brainpicarray' which is the floating point data array of the image we want to equalize  having the value between 0 and 1 . The  function returns the image after histogram equalization . Now, we can plot the equalized image.

### Brief Explanation : In order to increase the contrast of a medical image,we use histogram equalization. Generally the image that is formed have two main tones i.e black and white . To distribute the pixels in other tones i.e. gray, we use histogram equalization. Thus the new image formed will be more detailed and thus more information can be fetched about the medical condtion of the subject. Basicallly its a technique which consists of adjusting the gray scale of the image so that the gray level histogram of the input image is mapped onto a uniform histogram . The histogram equalization technique is based on a transformation using the histogram of a complete image in histogram equalization, the goal is to obtain a uniform histogram for the output image.

### BRIEF EXPLANATION: The advantage of adaptive histogram equalization is that it splits the image into small rectangular areas called tiles, and enhances the contrast of these areas by adjusting their local histograms. This method is also known as contrast limited adaptive histogram equalization (CLAHE).
