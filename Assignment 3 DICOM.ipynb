```python
%matplotlib inline
from skimage import novice, data, io, exposure
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import dicom


Dfile = dicom.read_file('IM_0028.dcm')


ConstPixelDims = (int(Dfile.Rows), int(Dfile.Columns))
ArrayDicom = np.zeros(ConstPixelDims, dtype=Dfile.pixel_array.dtype)
ds = dicom.read_file('IM_0028.dcm')
ArrayDicom[:,:]=ds.pixel_array

plt.figure()     # We can enlarge the picture by changing the value of 'figsize' or 'dpi'(dots per inch)


fig=plt.figure(figsize=(100, 100), dpi=80, facecolor='w', edgecolor='k') #figsize changed as compare to the values examined in the clas
plt.axes().set_aspect('equal') 
plt.set_cmap(plt.gray()) 
x=np.arange(Dfile.Rows)
y=np.arange(Dfile.Columns)

plt.pcolormesh(x, y, np.flipud(ArrayDicom[:, :]))
plt.axis([x.min(),x.max(),y.min(),y.max()])
```

## ANSWER (f)


### Pixel Spacing(0028, 0030): It is the distance between the the centre of each pixel(while diagonising patient) . It is usually given in a numeric pair. One of them gives the adjacent row spacing and other the adjacent column spacing(unit : delimeter)

### Image Position(0020, 0032) : This attribute gives us the information about the x,y and z coordinate of the voxels in the image. (in milimeter)

### Image Orientation(0020, 0037) : It gives the mathematical orientation of the patient i.e. the direction cosines of the row and column with respect to the orientation of the patient.

### Slice Thickness (0018, 0050) : It is the thickness of the slice used to examine the modality . ( in milimeter)

### Slice Location(0020, 1041) : It is the relative position of the image place with respect to the postion of the reference plane.(in milimeter)
