# While scanning Tomato:
The MR sequence that was used to aquire the image was spin Echo 'T2' , the pulse sequence name was 'TSE'. The Echo pulse sequence was 'SPIN'. The MR sequence as we know is a specific timing scheme with series of different gradients,readout and excitations. The scanning sequence is 'SE' which stands for spin Echo ,which consists of two RF Pulses. One of which is an excitation(alpha) and the other is the aquisition pulse.

The two of the main parameters of the spin Echo sequence are TE (Echo time) i.e the time taken for the excitation pulse to reach the maximum value(echo maximum).
The other parameter is the TR ( Repetetion  time) i.e the time between two succesive excitation pulses. 
TR  should be long enough so  that every tissue undergo T1 relaxation.
Since, mostly only one line is aquired in one repetion time, thus TA( aquisition time) is directly propotional to the the repetetion time(TR).

# While scanning watermelon :
The scanning sequence that was the 'GR' i.e. Gradient Recalled . An excitation pulse of flip angle less than or equal to 90 degree is used.It does not use the 180 degree refocussing pulse. The advantage of using flip angle lower than 90 degree is it decreases the amount of magnetization directed in the transverse plane . Thus , we have a faster recovery of the longitudnal magnetization which results in shorter repetetion time(TR) and scanning time. Thus the aquisition time also becomes faster.







# (3) Difference between aquisition Matrix and pixels of the image


The aquisition matrix that i have got for one my images formed is  Acquisition Matrix :[0, 400, 319, 0] , which is different from the pixels of my picture i.e. 512 rows and 512 columns. Aquisition matrix is basically the number of independent data samples in every direction. It can have a different value in phase-encoding direction and a different value in frequency-encoding direction . While reconstruction the image the aquisition matrix can have different values in time and angle. Every time most probably the aquisition will be having a different value whereas the number of rows and columns (pixels) are a  fixed quantity . For example another image of the study is having the aquisition matrix of [0, 256, 128, 0] which is differnt from the pixel information i.e. 256 rows and 256 columns.
