By knowing this duration, we can know whether the signal is abnormal or normal8. As per the standards, the standard QRS duration of a normal ECG signal will rangefrom 0.06 to 0.10 seconds.
 
 
###### R-R interval: R-peak is the longest amplitude peak in ECG signal.  e R-R interval is calculated by dividing number of samples between two R peaks and sampling frequency of the signal. It plays a vital role in  nding abnormalities of a signal.
###### Heart rate: Heart rate can be measured by using the formulae shown in Equation down below

```python
%matplotlib inline

import scipy.linalg
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.pyplot as plt


q, a = signal.butter(4, 100./360)
w, h = signal.freqs(q, a)
b=signal.lfilter(q,a,AllSignal) #we filter the signal


plt.plot(w, 20 * np.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth frequency response')
plt.xlabel('Frequency  in radians / second')
plt.ylabel('Amplitude [dB]')

plt.margins(0,0.1)
plt.grid(which='both', axis='both')
plt.axvline(100, color='purple') # 100 Hz being the cutoff frequency
plt.show()
```

### Effect of the Filter-  

### Since our designed filter is a butterworth, we expect the passing band to be very smooth without any ripples . Thus the calculation of the R-R interval will not be a problem. Therefore, the calculation of the heart beat will become more easy and efficient . 


### It will  effect our calculation for heart rate as we have designed the filter for cutoff of 100 Hz . Thus , any freqency component over 100Hz gets attenuated which might hinder the calculation of heart rate  . The PQ and ST segments can also be effected .  

### Thus we can see , with the increase in order of the butterworth filter the  downward slope towards cutoff frequency gets steeper and steeper.
 
### If no frequency component is above the cutoff frequency , then the filtering of the signal can be very benificial. Resulting is almost accurate calculation of the heart rate.
